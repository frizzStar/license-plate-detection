```python
import cv2
import os
```

```python
# ===CONFIGURATION===
video_path = "C:/Users/fahri/OneDrive/Documents/Skripsi/data/raw/data_21_april_2024_gedel/video3.mp4"
output_folder = "C:/Users/fahri/OneDrive/Documents/Skripsi/data/extracted_frames/data_21_april_2024/extracted_frames3_(3fps)"
frame_skip = 3
```

```python
# (Optional) Create Output Folder
os.makedirs(output_folder, exist_ok=True)
```

```python
# Open Video File
cap = cv2.VideoCapture(video_path)
if not cap.isOpened():
    raise FileNotFoundError(f"Could not open video: {video_path}")

count = 0 
saved_count = 0

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break # End of video

    # Save every 'frame_skip' frame
    if count % frame_skip == 0:
        frame_filename = os.path.join(output_folder, f"frame_{saved_count:04d}.jpg")
        cv2.imwrite(frame_filename, frame)
        saved_count += 1
        print(f"Saved: {frame_filename}")

    count += 1

cap.release()
print(f"Done! Extracted {saved_count} frames to {output_folder}.")
```
